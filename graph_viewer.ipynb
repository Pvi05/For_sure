```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout

G = nx.read_graphml("output/20260222_112945/provenance.graphml")

pos = graphviz_layout(G, prog="dot")

# Use d1 (author handle) as label, fallback to node id if missing
labels = {
    n: f"{G.nodes[n].get('author_handle', '?')}\n{G.nodes[n].get('date', '')[:10]}"
    for n in G.nodes
}

node_id, attrs = list(G.nodes(data=True))[0]
print(node_id)
print(attrs)


plt.figure(figsize=(14, 10))
nx.draw(
    G, pos,
    labels=labels,          # ← custom labels
    with_labels=True,
    arrows=True,
    node_color="lightblue",
    node_size=800,
    font_size=7,
    arrowsize=15,
    edge_color="gray",
    width=0.8,
)
plt.tight_layout()
plt.show()
```
